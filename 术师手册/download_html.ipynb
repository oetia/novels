```python
import time

for idx in range(len(links)):
    link = links[idx]
    filename = link.text
    url = f"https://www.piaotia.com/html/13/13393/{link.get('href')}"
    print(idx, filename, url)
    response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
    
    if response.status_code == 200:
        with open(f"./html/{idx}--*--{filename}.html", "wb") as dest:
            dest.write(response.content)
    else:
        break
    
    time.sleep(0.25)
```

```python
import os
for idx, link in enumerate(links):
    filename = f"{link.text}.html"
    filepath = f"./html/{filename}"
    os.rename(filepath, f"./html/{idx}---{filename}.html")
```
