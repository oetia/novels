```python
from bs4 import BeautifulSoup
import os
```

```python
ordered = list(sorted(os.listdir("./html"), key=lambda x: int(x.split("--*--")[0])))
```

```python
bs = BeautifulSoup(open(f"./html/{ordered[1089]}", "r", encoding="gbk").read(), "html.parser")

# remove script tags from bs
for script in bs(["script", "style", "meta", "div", "table"]):
    script.extract()
    

print(bs.prettify())
```
