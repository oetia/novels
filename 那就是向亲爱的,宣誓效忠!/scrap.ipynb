```python
import requests
import bs4
from bs4 import BeautifulSoup
```

```python
URL = "https://www.trxs.cc/tongren/7106/1.html"

res = requests.get(URL)
soup = BeautifulSoup(res.content.decode("gbk"), "html.parser")
```

```python
header = soup.find("div", class_=["read_chapterName"]).find('h1')
content = soup.find("div", class_=["read_chapterDetail"]).find_all("p")
```

```python
ps = list(map(lambda tag: tag.text, content))
ps_joined = "\n".join(ps)
```

```python
# CN SCRAP

with open("out.txt", "a+") as out_file:
    for index in range(500, 540):
        URL = f"https://www.trxs.cc/tongren/7106/{index+1}.html"
        res = requests.get(URL)
        soup = BeautifulSoup(res.content.decode("gbk"), "html.parser")
        
        header = soup.find("div", class_=["read_chapterName"])
        title = header.find("h1").text
        
        content = soup.find("div", class_=["read_chapterDetail"])
        ps = list(map(lambda tag: tag.text, content))
        ps_joined = "\n".join(ps)
        
        out_file.write(f"============================\n{title}\n============================\n")
        out_file.write(f"{ps_joined}\n")
        
```

```python
# getting chapter list
res = requests.get("https://hostednovel.com/novel/swear-fealty-to-me-my-subjects#chapters")
soup = BeautifulSoup(res.text, "html.parser")
```

```python
chaps = soup.find("ul", attrs={"role": "list"}, class_=["mt-6","pb-5"]).find_all("li")
```

```python
chap_links = list(map(lambda chap: chap.find("a").get("href"), chaps))
```

```python
with open("out.txt", "a+") as out_file:
    for link in chap_links:
        res = requests.get(link)
        soup = BeautifulSoup(res.text, "html.parser")
        content = soup.find("div", id="chapter-content").text
        
        out_file.write(content)
```
